```python
import pandas as pd
import numpy as np
import cv2
from cvtools import ipynb_show_color_histogram
from cvtools import ipynb_show_cv2_image
from cvtools import ipynb_show_matrix


# In[]
def splitFrame(array, gridSize):
    r, h = array.shape
    return (array.reshape(h//gridSize, gridSize, -1, gridSize)
                 .swapaxes(1, 2)
                 .reshape(-1, gridSize, gridSize))


def extract_frame_hs_histogram(frame):
  histogram = []
  hsv_image = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
  range = [[0, 180], [0, 256]]
  for i in [0,1]:
    channel = frame[:,:,i]
    channel_range = range[i]


    channel_histogram = cv2.calcHist(
        [hsv_image],
        [i],
        None, # one could specify an optional mask here (we don't use this here),
        [8],
        channel_range
        )
    
    histogram = np.append(histogram, channel_histogram)

  return histogram / np.sum(histogram)  # Return a normalized histogram.


GRID_SIZE = 2

def pipeline(frame, name):
    subframes = splitFrame(frame, GRID_SIZE)

    full_hist = extract_frame_hs_histogram(frame)
    np.save(full_hist, name)
    for i in enumerate(GRID_SIZE**2):
        sub_hist = extract_frame_hs_histogram(subframes[i])
        np.save(sub_hist, '{}_{}_{}'.format(name, i, i%GRID_SIZE))



#%%
img = cv2.imread('clouds-country-daylight-371633.jpg')
print(np.shape(img))
ipynb_show_cv2_image(img)
#%%
```
